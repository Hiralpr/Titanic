```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
t= sns.load_dataset("titanic")
```

```python
t
```

```python
t.isnull().sum()
```

```python
mean_age = t['age'].mean()
```

```python
mean_age
```

```python
t.age.fillna(mean_age, inplace = True)
```

```python
t.isnull().sum()
```

```python
t.deck.unique()

```

```python
mode_deck = t['deck'].mode()
```

```python
mode_deck
```

```python
t.deck.fillna(mode_deck, inplace = True)
#t.deck.fillna(mode_deck, inplace = True)
```

```python
t.isnull().sum()
```

```python
t['deck'] =t['deck'].fillna(t['deck'].mode()[0])
```

```python
t = t.dropna()
```
